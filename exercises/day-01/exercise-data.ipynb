```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./wdi-data-exercise-day01-all.csv")
```

```python
data.head()
```

```python
data['Country Code'].unique()
```

```python
data['Series Name'].unique()
```

```python
AUS = data[data['Country Code'] == 'AUS'].copy()
```

```python
age_brackets = []
for sn in AUS['Series Name']:
    a,b = sn.split(",")
    a = a.replace("Population ages ", "")
    if a == '80 and above':
        a = '80+'
    age_brackets.append(a)
```

```python
AUS['age'] = age_brackets
```

```python
AUS.head()
```

```python
gender = []
for sn in AUS['Series Name']:
    if 'female' in sn:
        gender.append('female')
    elif 'male' in sn:
        gender.append('male')
    else:
        print("cannot determine gender -- check data source")
```

```python
AUS['gender'] = gender
```

```python
AUS.head()
```

```python
plot_data = AUS[['1960', '2023', 'age', 'gender']].set_index(['age','gender'])
```

```python
plot_data
```

```python
plot_data = plot_data.unstack()
```

```python
plot_data.head()
```

```python
plot_data[("1960", "female")]
```

```python
plot_data[('1960', 'female')] = plot_data[('1960', 'female')] * -1
```

```python
plot_data[('2023', 'female')] = plot_data[('2023', 'female')] * -1
```

```python
plot_data
```

```python
pdata_1960 = plot_data["1960"].reset_index()
```

```python
pdata_1960.columns.name=""
```

```python
import seaborn as sns
ax1 = sns.barplot(x='male', y='age', data=pdata_1960, color='blue')
ax2 = sns.barplot(x='female', y='age', data=pdata_1960, color='green')
plt.title("Population pyramid for Australia, 1960 estimates")
plt.xlabel("% of population\nFemale / Male")
plt.grid()
```
